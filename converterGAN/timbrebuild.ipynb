```python
#CQTRainbowgram Transform
class RainbowgramCQT:
    def __init__(self, sr=22050, hop_length=512, n_bins=84):
        self.sr = sr
        self.hop_length = hop_length
        self.n_bins = n_bins

    def transform(self, folder_path, output_folder):
        # Create the output folder if it doesn't exist
        os.makedirs(output_folder, exist_ok=True)

        # Get the list of audio files in the input folder
        audio_files = [file for file in os.listdir(folder_path) if file.endswith('.wav')]

        for file in audio_files:
            # Load the audio file
            audio, _ = librosa.load(os.path.join(folder_path, file), sr=self.sr)

            # Compute the CQT spectrogram
            cqt = librosa.cqt(audio, sr=self.sr, hop_length=self.hop_length, n_bins=self.n_bins)

            # Normalize the CQT spectrogram
            cqt_norm = librosa.amplitude_to_db(np.abs(cqt), ref=np.max)

            # Compute the rainbowgram visualization
            rainbowgram = librosa.display.specshow(cqt_norm, sr=self.sr, hop_length=self.hop_length,
                                                   cmap='rainbow')

            # Save the rainbowgram as an image
            output_path = os.path.join(output_folder, f"{os.path.splitext(file)[0]}.jpg")


            plt.savefig(output_path)
            plt.close()

            print(f"Saved rainbowgram CQT for {file} as {output_path}")
```

```python
rainbowgram_cqt = RainbowgramCQT(sr=22050, hop_length=512, n_bins=84)
```
